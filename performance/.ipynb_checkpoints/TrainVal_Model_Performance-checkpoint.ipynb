```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Model Results

```python
initial_baseline = pd.read_csv('history_CNN_VanillaRNN.csv')
initial_baseline
```

```python
improved_baseline = pd.read_csv('history_M0_CNNImproved_VanillaRNN.csv')
improved_baseline
```

```python
initial_m1 = pd.read_csv('history_CNN_GRU.csv')
initial_m1
```

```python
improved_m1 = pd.read_csv('history_M1_CNNImproved_GRU.csv')
improved_m1
```

```python
initial_m2 = pd.read_csv('history_CNN_LSTM.csv')
initial_m2
```

```python
improved_m2 = pd.read_csv('history_M2_CNNImproved_LSTM.csv')
improved_m2
```

```python
final_model = pd.read_csv('history_MobileNetV2_LSTM.csv')
final_model
```

Summarized Train Accuracy

```python
train_summary = {"Model": ["CNN + RNN (Baseline)", 
                           "CNN + GRU", 
                           "CNN + LSTM",
                           "CNN + RNN (Improved)",
                           "CNN + GRU (Improved)",
                           "CNN + LSTM (Improved)", 
                           "MobileNetV2 + LSTM"],
                "Train Accuracy": [initial_baseline.iloc[19, 2],
                                  initial_m1.iloc[19, 2],
                                  initial_m2.iloc[19, 2],
                                  improved_baseline.iloc[19, 2],
                                  improved_m1.iloc[19, 2],
                                  improved_m2.iloc[19, 2],
                                  final_model.iloc[19, 2]] 
                }

pd.DataFrame(train_summary)
```

# Accuracy Graph

## Consolidated Models: Initial Performance 

```python
# Baseline Initial
plt.plot(initial_baseline['epoch'], initial_baseline['train_acc'], color='blue', label='CNN + RNN (Baseline)')
plt.plot(initial_baseline['epoch'], initial_baseline['val_acc'], linestyle="--", color='blue', alpha=0.5)

# CNN + GUR Initial
plt.plot(initial_m1['epoch'], initial_m1['train_acc'], color='darkred', label='CNN + GRU')
plt.plot(initial_m1['epoch'], initial_m1['val_acc'], linestyle="--", color='darkred', alpha=0.5)

# CNN + LSTM Initial
plt.plot(initial_m2['epoch'], initial_m2['train_acc'], color='goldenrod', label='CNN + LSTM')
plt.plot(initial_m2['epoch'], initial_m2['val_acc'], linestyle="--", color='goldenrod', alpha=0.5)

plt.xticks(range(1, 21))
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.title("Model Performance (Initial Models)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Baseline Model: Initial vs Improved Performance

```python
# Baseline Initial
plt.plot(initial_baseline['epoch'], initial_baseline['train_acc'], color='gray', label='CNN + RNN (Baseline) - Initial')
plt.plot(initial_baseline['epoch'], initial_baseline['val_acc'], linestyle="--", color='gray', alpha=0.5)

# Baseline Improved
plt.plot(improved_baseline['epoch'], improved_baseline['train_acc'], color='blue', label='CNN + RNN (Baseline) - Improved')
plt.plot(improved_baseline['epoch'], improved_baseline['val_acc'], linestyle="--", color='blue', alpha=0.5)

plt.xticks(range(1, 21))
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.title("Model Performance")
plt.legend(loc='upper left', bbox_to_anchor=(-0.05, -0.15))
plt.show()
```

Note: broken lines are validation scores, while solid lines are train scores

## Consolidated Models: Initial vs Improved Performance

```python
# Baseline Initial
plt.plot(initial_baseline['epoch'], initial_baseline['train_acc'], color='blue', label='CNN + RNN (Baseline)')
plt.plot(initial_baseline['epoch'], initial_baseline['val_acc'], linestyle="--", color='blue', alpha=0.5)

# CNN + GUR Initial
plt.plot(initial_m1['epoch'], initial_m1['train_acc'], color='red', label='CNN + GRU')
plt.plot(initial_m1['epoch'], initial_m1['val_acc'], linestyle="--", color='red', alpha=0.5)

# CNN + LSTM Initial
plt.plot(initial_m2['epoch'], initial_m2['train_acc'], color='goldenrod', label='CNN + LSTM')
plt.plot(initial_m2['epoch'], initial_m2['val_acc'], linestyle="--", color='goldenrod', alpha=0.5)

# Final Model
plt.plot(final_model['epoch'], final_model['train_acc'], color='green', label='MobileNetV2 + LSTM')
plt.plot(final_model['epoch'], final_model['val_acc'], linestyle="--", color='green', alpha=0.5)

# Baseline Improved
plt.plot(improved_baseline['epoch'], improved_baseline['train_acc'], color='blue', label='CNN + RNN - Improved')
plt.plot(improved_baseline['epoch'], improved_baseline['val_acc'], linestyle="--", color='blue', alpha=0.5)

# CNN + GRU Improved
plt.plot(improved_m1['epoch'], improved_m1['train_acc'], color='red', label='CNN + GRU - Improved')
plt.plot(improved_m1['epoch'], improved_m1['val_acc'], linestyle="--", color='red', alpha=0.5)

# CNN + LSTM Improved
plt.plot(improved_m2['epoch'], improved_m2['train_acc'], color='goldenrod', label='CNN + LSTM - Improved')
plt.plot(improved_m2['epoch'], improved_m2['val_acc'], linestyle="--", color='goldenrod', alpha=0.5)

plt.xticks(range(1, 21))
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.title("Training Performance per Model")
plt.legend(loc='upper left', bbox_to_anchor=(-0.05, -0.15), ncols=2)
plt.show()
```
